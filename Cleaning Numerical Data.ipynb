```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn
```

```python
customer_df = pd.read_csv("https://raw.githubusercontent.com/ironhack-labs/lab-cleaning-numerical-data/master/files_for_lab/we_fn_use_c_marketing_customer_value_analysis.csv")
```

```python
#Renaming the columns so they follow the PE8 (snake case)
#Putting an underscore where theres a space
customer_df.columns = customer_df.columns.str.replace(' ', '_')
 
```

```python
customer_df
```

```python
#dropping the column customer -- not necessary 
customer_df = customer_df.drop(columns=['Customer'])
```

```python
numeric_data = customer_df.select_dtypes(include=np.number).columns
numeric_data
```

```python
#Plotting a correlation matrix

correlation = customer_df.select_dtypes(include=np.number).corr()
correlation
```

There's correlation between Total Claim Amount and Monthly Premium Auto, Income

```python
#Let's tale a look at the correlation with a heatmap

sns.heatmap(correlation, annot=True)
```

Lets plot the continous variables

```python
for i in ['Customer_Lifetime_Value', 'Income', 'Monthly_Premium_Auto',
       'Months_Since_Last_Claim', 'Months_Since_Policy_Inception',
       'Number_of_Open_Complaints', 'Number_of_Policies']: 
    sns.scatterplot(data = customer_df, x = i, y = 'Total_Claim_Amount')
    plt.show()
```

Let us plot categorical variables using a histogram in order to see the distribution of the variables. 

```python
categorical_data = customer_df.select_dtypes(include='object')
categorical_data.columns
```

```python
for i in categorical_data: 
    sns.histplot(data=customer_df, x=i)
    plt.show()
```

Comments on the plots above: 
Most of the vehicles are medium-sized. 
The most common vehicle-class is the Four-Door car
The most-used Sales Channel is the Agent, followed by the Branch and Call-Center. The least used is the Web.
The policy type with more more customers is Personal Auto. Special Auto has the least. 
The highest numnber of customers are married. Most of them are employed and live in Sub-Urban areas.
